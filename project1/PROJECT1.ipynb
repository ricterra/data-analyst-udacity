```python
global_temp = pd.read_csv('global_temp.csv')
```

```python
global_temp
```

```python
norm_global_temp = global_temp.avg_temp/global_temp.avg_temp.iloc[0]
```

```python
norm_global_temp
```

```python
plt.plot(global_temp.year, global_temp.avg_temp.rolling(window=1).mean(),label='MA 1 year')
plt.plot(global_temp.year, global_temp.avg_temp.rolling(window=5).mean(),label='MA 5 year')
plt.plot(global_temp.year, global_temp.avg_temp.rolling(window=10).mean(),label='MA 10 year')
plt.xlabel('years')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Global Average Temperature')
plt.legend(loc='best')
plt.savefig('global_temp_MA.png')
plt.show()
```

```python
plt.plot(global_temp.year, global_temp.avg_temp.rolling(min_periods=1, center=True, window=10).mean(),label='MA 10 year')
plt.xlabel('years')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Global Average Temperature')
plt.legend(loc='best')
plt.show()
```

```python
norm_global_temp.rolling(min_periods=1, center=True, window=10).mean()
```

```python
plt.plot(global_temp.year, norm_global_temp.rolling(min_periods=1, center=True, window=10).mean()*100,label='MA 10 year')
plt.xlabel('years')
plt.ylabel('Temperature Growth (%)')
plt.title('Global Average Temperature')
plt.legend(loc='best')
plt.show()
```

```python
city_temp = pd.read_csv('city_temp.csv')
```

```python
norm_city_temp = city_temp.avg_temp/city_temp.avg_temp.iloc[0]
```

```python
norm_city_temp
```

```python
norm_city_temp.rolling(min_periods=1, center=True, window=10).mean()
```

```python
plt.plot(city_temp.year, city_temp.avg_temp.rolling(min_periods=1, center=True, window=1).mean(),label='MA 1 year')
plt.plot(city_temp.year, city_temp.avg_temp.rolling(min_periods=1, center=True, window=5).mean(),label='MA 5 year')
plt.plot(city_temp.year, city_temp.avg_temp.rolling(min_periods=1, center=True, window=10).mean(),label='MA 10 year')
plt.xlabel('years')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Rio de Janeiro Average Temperature')
plt.legend(loc='best')
plt.savefig('city_temp_MA.png')
plt.show()
```

```python
plt.plot(city_temp.year, city_temp.avg_temp.rolling(min_periods=1, center=True, window=10).mean(),label='MA 10 year')
plt.xlabel('years')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Rio de Janeiro Average Temperature')
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(city_temp.year, norm_city_temp.rolling(min_periods=1, center=True, window=10).mean()*100,label='MA 10 year')
plt.xlabel('years')
plt.ylabel('Temperature Growth (%)')
plt.title('Rio de Janeiro Average Temperature')
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(global_temp.year, global_temp.avg_temp.rolling(min_periods=1, center=True, window=10).mean(),label='Global MA 10 year')
plt.plot(city_temp.year, city_temp.avg_temp.rolling(min_periods=1, center=True, window=10).mean(),label='Rio MA 10 year')
plt.xlabel('years')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Global vs Rio de Janeiro Avg Temperature')
plt.legend(loc='best')
plt.savefig('global_city_noscale.png')
plt.show()
```

```python
plt.plot(global_temp.year, norm_global_temp.rolling(min_periods=1, center=True, window=10).mean()*100,label='Global MA 10 year')
plt.plot(city_temp.year, norm_city_temp.rolling(min_periods=1, center=True, window=10).mean()*100,label='Rio MA 10 year')
plt.xlabel('years')
plt.ylabel('Temperature Growth (%)')
plt.title('Global vs Rio de Janeiro Temperature Growth')
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(global_temp.year, norm_global_temp.rolling(min_periods=1, center=True, window=10).mean()*100,label='Global MA 10 year')
plt.plot(city_temp.year, norm_city_temp.rolling(min_periods=1, center=True, window=10).mean()*100,label='Rio MA 10 year')
plt.xlabel('years')
plt.ylabel('Temperature Growth (%)')
plt.title('Global vs Rio de Janeiro Temperature Growth')
plt.legend(loc='best')
plt.xlim(1820,2020)
plt.grid(linestyle='dotted')
plt.savefig('global_city_scale.png')
plt.show()
```

```python
ratio = norm_global_temp.rolling(min_periods=1, center=True, window=10).mean()/norm_city_temp.rolling(min_periods=1, center=True, window=10).mean()
```

```python
ratio
```

```python
plt.plot(global_temp.year, ratio*100,label='Ratio MA 10 year')
plt.xlabel('years')
plt.ylabel('Relative Temperature Growth (%)')
plt.title('Ratio Global/Rio Avg Temperature')
plt.legend(loc='best')
plt.xlim(1820,2020)
plt.grid(linestyle='dotted')
plt.savefig('plot2.png')
plt.show()
```
